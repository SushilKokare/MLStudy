## Data Science

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
ohe = OneHotEncoder(sparse=False)
```

```python
ohe.fit_transform(df[['CourseCategory']])
```

```python
newdf = pd.get_dummies(df['CourseCategory'])
```

```python
df = pd.concat((df,newdf),axis=1)
df.head()
```

```python
df = df.drop(['CourseCategory'],axis=1)
df.head()
```

```python
x = df.drop(['CourseCompletion','UserID'],axis=1)
y = df[['CourseCompletion']]
```

```python
xtr, xts, ytr, yts = train_test_split(x,y,test_size=0.25,random_state=43)
```

```python
xtr.shape, xts.shape, ytr.shape, yts.shape
```

```python
results = dict()
```

```python
models = {
    'LR' : LogisticRegression(),
    'DTC' : DecisionTreeClassifier(),
    'SVC' : SVC(),
    'KNNC' : KNeighborsClassifier(),
    'BNB' : BernoulliNB(),
    'RFC' : RandomForestClassifier()
}
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
for name,model in models.items():
    print(name)
    model.fit(xtr,ytr)
    y_pred = model.predict(xts)
    acc = accuracy_score(yts,y_pred)
    results[name] = acc
```

```python
results
```

```python
result = pd.DataFrame( {'Model' : results.keys(),
            'Acc' : np.array(list(results.values()))*100})
```
